```python
def det_only_char(str):
    acc_list=['q','w','e','r','t','y','u','i','o','p','a','s','d','f','g','h','j','k','l','z','x','c','v','b','n','m','Q','W','E','R','T','Y','U','I','O','P','A','S','D','F','G','H','J','K','L','Z','X','C','V','B','N','M']
    for c in str:
        if c not in acc_list:
            return False
    return True
```

```python
from scipy.stats import chisquare
import math
```

```python
not_eq_words=[]
eq_words=[]
word_and_p=[]

save_mbti_pv={}
pv_by_mbti={}

all_word_and_pv=[]

for m in all_mbti:
    save_mbti_pv[m]=[]

for word_index in range(whole_word_num):
    now_o=[]
    now_e=[]
    that_word_total=0
    all_word_total=0
    for m in all_mbti:
        that_word_total+=word_by_type[m][word_index]
        all_word_total+=type_total[m]
    
    pv_by_mbti[word_by_index[word_index]]={}
    
    for m in all_mbti:
        oi = word_by_type[m][word_index]
        ei = that_word_total*(type_total[m]/all_word_total)
        now_o.append(oi)
        now_e.append(ei)
        save_mbti_pv[m].append( [(oi-ei)/math.sqrt(ei),word_by_index[word_index]] )
        pv_by_mbti[word_by_index[word_index]][m]=[]
        pv_by_mbti[word_by_index[word_index]][m].append((oi-ei)/math.sqrt(ei))
            
    va = chisquare(now_o, f_exp=now_e).pvalue
    
    all_word_and_pv.append([chisquare(now_o, f_exp=now_e).statistic,word_by_index[word_index]])
    
    
    if va<0.05:
        not_eq_words.append(word_by_index[word_index])
    else:
        eq_words.append(word_by_index[word_index])
```
